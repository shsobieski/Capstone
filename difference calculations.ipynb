```python
import pandas as pd
```

```python
RvD = pd.read_excel('data/predictions/diffs/RvD.xlsx', index_col=0)
RvD['median']= RvD.median(axis=1)
RvD['mean']= RvD.mean(axis=1)
RvD['median'].plot()
RvD['mean'].plot();
```

```python
RvI = pd.read_excel('data/predictions/diffs/RvI.xlsx', index_col=0)
RvI['median']= RvI.median(axis=1)
RvI['mean']= RvI.mean(axis=1)
RvI['median'].plot()
RvI['mean'].plot();
```

```python
DvI = pd.read_excel('data/predictions/diffs/DvI.xlsx', index_col=0)
DvI['median']= DvI.median(axis=1)
DvI['mean']= DvI.mean(axis=1)
DvI['median'].plot()
DvI['mean'].plot();
```

```python
RvD.to_excel('data/predictions/diffs/RepvDem.xlsx')
RvI.to_excel('data/predictions/diffs/RepvInd.xlsx')
DvI.to_excel('data/predictions/diffs/DemvInd.xlsx')
```
